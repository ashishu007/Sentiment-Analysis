```python
import pandas as pd
import re
df = pd.read_csv("C:\\Users\\User\\Documents\\Sentiment Analysis\\Resources\\trainingandtestdata\\testing.csv", parse_dates=["time"])
df.head(10)
```

```python
df = df.drop(['useless', 'useless1', 'id', 'time'], axis=1)
df.head(10)
```

```python
for i in range(len(df["text"])):
    temp = df['text'][i]
    temp1 = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",temp).split())
    temp2 = temp1.replace("RT", "")
    df['text'][i] = temp2

df.head(10)
```

```python
df.to_csv("C:\\Users\\User\\Documents\\Sentiment Analysis\\Resources\\trainingandtestdata\\testing_final.csv")
```
